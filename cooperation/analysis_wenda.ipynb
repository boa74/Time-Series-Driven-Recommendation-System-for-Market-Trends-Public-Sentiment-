### Turn state level weather into national level

```python
pop = pd.read_excel("apportionment-2020-table02.xlsx")

print(pop.columns)
```

```python
state_col = "U.S. Department of Commerce"
pop_col = "Unnamed: 1"

pop_clean = pop[pd.to_numeric(pop[pop_col], errors="coerce").notnull()].copy()

pop_clean = pop_clean.rename(columns={
    state_col: "state",
    pop_col: "population"
})

pop_clean = pop_clean[~pop_clean["state"].isin([
    "TOTAL RESIDENT POPULATION1",
    "TOTAL RESIDENT POPULATION, INCLUDING PUERTO RICO",
    "Puerto Rico"
])]

pop_clean["population"] = pop_clean["population"].astype(int)

total_pop = pop_clean["population"].sum()
pop_clean["weight"] = pop_clean["population"] / total_pop

if "Unnamed: 2" in pop_clean.columns:
    pop_clean = pop_clean.drop(columns=["Unnamed: 2"])

pop_clean = pop_clean.reset_index(drop=True)

pop_clean
```

```python
# --- CLEAN bad columns ---
weather_data = weather_data[[c for c in weather_data.columns 
                             if not c.startswith("temp_mean_")
                             and c != "date_Rainfall"
                             and not c.endswith("_w")
                             and not c.startswith("National")]]

# Convert temps to numeric
temp_cols = [c for c in weather_data.columns if "_Temp_Max" in c or "_Temp_Min" in c]
for col in temp_cols:
    weather_data[col] = pd.to_numeric(weather_data[col], errors="coerce")

# Correct state extractor
def get_state(col):
    return (col.replace("_Temp_Max","")
               .replace("_Temp_Min","")
               .replace("_Rainfall",""))

valid_states = set(pop_clean["state"])
state_weights = pop_clean.set_index("state")["weight"].to_dict()

# Identify columns correctly
rain_cols = [c for c in weather_data.columns if c.endswith("_Rainfall") and get_state(c) in valid_states]
tmax_cols = [c for c in weather_data.columns if c.endswith("_Temp_Max") and get_state(c) in valid_states]
tmin_cols = [c for c in weather_data.columns if c.endswith("_Temp_Min") and get_state(c) in valid_states]

# National rainfall
weather_data["National_Rainfall"] = sum(weather_data[c] * state_weights[get_state(c)] for c in rain_cols)

# National temperature max
weather_data["National_Temp_Max"] = sum(weather_data[c] * state_weights[get_state(c)] for c in tmax_cols)

# National temperature min
weather_data["National_Temp_Min"] = sum(weather_data[c] * state_weights[get_state(c)] for c in tmin_cols)

# Final dataset
national_weather = weather_data[[
    "Date",
    "National_Rainfall",
    "National_Temp_Max",
    "National_Temp_Min"
]].copy().sort_values("Date").reset_index(drop=True)

national_weather
```

# Volume to return

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# -------------------------------------------------------------
# 1. Create analysis3 dataset
# -------------------------------------------------------------
analysis3 = analysis2.copy().sort_values("Date")

# Create returns and volume changes
analysis3["S&P_return"] = np.log(analysis3["SP500_Close"]).diff()
analysis3["Volume_diff"] = analysis3["SP500_Volume"].diff()

# Remove rows with NA after differencing
analysis3 = analysis3.dropna(subset=["S&P_return", "Volume_diff"])

# -------------------------------------------------------------
# 2. Model A: OLS regression (Volume_diff → S&P_return)
# -------------------------------------------------------------
X_ols = sm.add_constant(analysis3["Volume_diff"])
y = analysis3["S&P_return"]

ols_model = sm.OLS(y, X_ols).fit()
print("\n===== OLS: Volume_diff → S&P_return =====")
print(ols_model.summary())

# -------------------------------------------------------------
# 3. Model B: ARIMAX (Volume_diff as exogenous)
# -------------------------------------------------------------
arimax_model = sm.tsa.SARIMAX(
    y,
    exog=analysis3[["Volume_diff"]],
    order=(1,0,1),        # classic specification for returns
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

print("\n===== ARIMAX: Volume_diff → S&P_return =====")
print(arimax_model.summary())

# -------------------------------------------------------------
# 4. Plot: Volume_diff and S&P_return
# -------------------------------------------------------------
plt.figure(figsize=(14,6))
plt.subplot(2,1,1)
plt.plot(analysis3["Date"], analysis3["Volume_diff"])
plt.title("Daily Change in Trading Volume (Volume_diff)")
plt.subplot(2,1,2)
plt.plot(analysis3["Date"], analysis3["S&P_return"])
plt.title("S&P Daily Log Return")
plt.tight_layout()
plt.show()
```
